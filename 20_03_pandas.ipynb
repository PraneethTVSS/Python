```python
# Pandas for data size
import pandas as pd
s = pd.Series([1, 2, 3, 4, 5])
print(s)
print('-------------')
print(s.values)
print('-------------')
print(s.index)
print('-------------')
print(s[1])
print('-------------')
print(s[1:2])
print('-------------')
print(s[1:-1])
```

```python
s = pd.Series([33, 545, 67, 32], index=['t', 'v', 's', 's'])
print(s)
print('--------------')
print(s['v'])
```

```python
_dict = {
    "name": "praneeth",
    "initial": "tvss",
    "city": "eluru",
    "state": "A.P."
}
s = pd.Series(_dict)
print(s)
print('-----------')
# Passing values as keys and keys as values

print('tried below')
print('Why is this happening ?')
s = pd.Series(_dict.keys(), index=_dict.values())
print(s)
print(type(_dict.keys()))
print('-------------')

print('Working Solution')
s = pd.Series([val for val in _dict.keys()], index=_dict.values())
print(s)

print('-----------')
```

```python
# Creation of DataFrames

cars = pd.Series(['swift', 'xuv', '5s', 'q7'], index=['maruti', 'mahindra', 'bmw', 'audi'])
print(cars)

print('---------------')
profits = pd.Series([100, 200, 500, 300], index=['maruti', 'mahindra', 'bmw', 'audi'])
print(profits)

print('---------------')
final = pd.DataFrame({'cars': cars, 'profits': profits})
final
```
